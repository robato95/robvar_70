# Pi Estimation Using Monte Carlo Simulations

```python
import random
import matplotlib.pyplot as plt
```

### Concept:
- Idea is to reandomly generate points inside a square and determine how many fall inside a circle inscribed within that square. The ratio of the points inside the circle to the total points generated can be used to estimate <strong>pi</strong>.

### Define the square and circle:
- The square has a side length of 2, centered at the origin - meaning the coordinates are (-1,-1) to (1,1).
- The circle has a radius of 1, centered at the origin, thus the equation is x^2 + y^2 <= 1.

### Calculation:
- The ratio of the points inside the circle to the total points is approx. equal to the ratio of the are of the circle to the area of the square.
- Since the area of the circle is <strong>pi</strong> x 1<sup>2</sup>, then the ratio is approximately <strong>pi</strong>&divide;4.


```python
# Create function that measures circle area
def circle_area_argument(x,y):
    if (x**2 + y**2) <= 1:
        return True
    else:
        return False 
        
# begin with 1 total points
total_points = 1000
inside_circle = 0
results = []

while total_points < 50001:
    for point in range(total_points):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if circle_area_argument(x,y) is True:
            inside_circle += 1
    pi = round(inside_circle/total_points*4,5)
    results.append((total_points, pi))
    inside_circle = 0
    total_points +=50
        
```

### Results
Based on our simulation parameters - I began the exercise with 1,000 points and chose to increase the number of points by 50 for each following simulation, until we've reached 50,0000 points. Therefore, we iterated through this exercise 981 times, totaling 249,255,000 points.

```python
# Plot Results
x_values = [total for total, _ in results]
y_values = [pi for _, pi in results]

plt.scatter(x_values, y_values, label = 'Estimation', color = 'green', marker = 'o', alpha = .2)
plt.axhline(y=3.14159, color='purple', linestyle='--', label="Actual Pi")
plt.xlabel("Total Points")
plt.ylabel("Estimated Pi")
plt.title("Estimation of Pi using Monte Carlo Simulation")
plt.legend()
plt.grid(True)
plt.show()
       
```

```python
# Error Terms
error_values = [(pi-3.14159) for _, pi in results]

plt.scatter(x_values, error_values, label = 'Error Term',color = 'red', marker = 'o',alpha=0.2)
plt.xlabel("Total Points")
plt.axhline(y=0, color='green', linestyle='--', label="Origin")
plt.ylabel("Error Term")
plt.title("Error Terms Plotted Against Number of Points Used")
plt.legend()
plt.grid(True)
plt.show()
```
